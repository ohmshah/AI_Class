```python
graph = {'A': set(['B', 'C']),
         'B': set(['A', 'C', 'D']),
         'C': set(['A', 'B', 'D', 'E']),
         'D': set(['B', 'C', 'E', 'F']),
         'E': set(['C', 'D', 'F']),
         'F': set(['D', 'E'])}

graph

def bfs(graph, start):
    
    visible = set()
    queue = [start]
    
    while queue:
        point = queue.pop(0)
        
        if point not in visible:
            visible.add(point)
            queue.extend(graph[point] - visible)
            
    return visible

bfs(graph, 'A')
```
