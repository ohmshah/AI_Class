```python
# BFS algorithm in Python
import collections

def bfs(graph, root):

    visited, queue = set(), collections.deque([root])
    visited.add(root) # Adding Root Node (Starting Node)

    while queue:

        node = queue.popleft()
        print(str(node) + " ", end="") # Format for outputting results

        for neighbour in graph[node]: # Checks each neighbour for each node
            if neighbour not in visited:
                visited.add(neighbour)
                queue.append(neighbour) # Adds neigbour to the queue


if __name__ == '__main__':
    graph = {'A': set(['B', 'C']),
         'B': set(['A', 'C', 'D']),
         'C': set(['A', 'B', 'D', 'E']),
         'D': set(['B', 'C', 'E', 'F']),
         'E': set(['C', 'D', 'F']),
         'F': set(['D', 'E'])}
    
    print("Following is Breadth First Traversal: ")
    bfs(graph, 'A')
```

```python
# DFS algorithm in Python
visited = set()    

def dfs(visited, graph, node):
    if node not in visited: # Checks if the current node is in the visited list
        print(str(node) + " ", end="") # Format for printing
        visited.add(node) # Adds the node to the visited list
        
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)

if __name__ == '__main__':
    graph = {'A': set(['B', 'C']),
         'B': set(['A', 'C', 'D']),
         'C': set(['A', 'B', 'D', 'E']),
         'D': set(['B', 'C', 'E', 'F']),
         'E': set(['C', 'D', 'F']),
         'F': set(['D', 'E'])}
    
    print("Following is Depth First Traversal: ")
    dfs(visited, graph, 'A')
```
